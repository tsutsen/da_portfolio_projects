## 1 Продукт

### 1.1 Количество пользователей в день/неделю/месяц

Для удобства создадим отдельную таблицу только с id пользователей и со временем начала и конца сессий

```python
user_activity = visits[['uid','start ts','end ts']].copy()

user_activity.columns = ['id','start','end']
user_activity.head()
```

Теперь посчитаем, в какой день, на какой неделе и в каком месяце состоялась сессия и добавим эти данные в таблицу

```python
user_activity['session_date'] = pd.to_datetime(user_activity['start'].dt.date)

user_activity['session_month'] = pd.to_datetime(user_activity['session_date'].astype('datetime64[M]'))

user_activity['session_week'] = pd.to_datetime(user_activity['session_date'], unit='d') - pd.to_timedelta(
    user_activity['session_date'].dt.dayofweek, unit='d')
```

Далее необходимо присвоить дату первой сессии каждому пользователю:


```python
first_session_date = user_activity.groupby(['id'])['session_date'].min()
first_session_date.name = 'first_session_date'
user_activity = user_activity.join(first_session_date,on='id')

user_activity['first_session_date'] = pd.to_datetime(user_activity['first_session_date'])
```

Первую неделю:

```python
user_activity['first_session_week'] = pd.to_datetime(
    user_activity['first_session_date'], unit='d') - pd.to_timedelta(
    user_activity['first_session_date'].dt.dayofweek, unit='d')
```

И первый месяц:

```python
user_activity['first_session_month'] = pd.to_datetime(
    user_activity['first_session_date'], unit='d') - pd.to_timedelta(
    user_activity['first_session_date'].dt.day -1, unit='d')
```

А сейчас можно расчитать количество активных пользователей в день, в неделю и в месяц:


```python
dau = int(user_activity.groupby('session_date').agg({'id': 'nunique'}).mean())
wau = int(user_activity.groupby('session_week').agg({'id': 'nunique'}).mean())
mau = int(user_activity.groupby('session_month').agg({'id': 'nunique'}).mean())
```

```python
print('Среднее количество пользователей\nВ день: {}\nВ неделю {}\nВ месяц: {}'.format(dau, wau, mau))
```

### 1.2  Количество сессий в день

Далее – подсчёт количества сессий в день

```python
sessions_per_day = user_activity.groupby(['session_date']).agg({'id': ['count','nunique']})
sessions_per_day.columns = ['sessions', 'users']
sessions_per_day['sessions_per_user'] = sessions_per_day['sessions'] / sessions_per_day['users']
sessions_per_day
```

```python
sessions_per_day['sessions_per_user'].mean()
```

В среднем одна сесссия в день. Нечасто.

### 1.3  Длительность сессий

Для расчёта длительности сессии вычтем время начала сессии из времени конца сессии. Для наглядности построим гистограмму и посмотрим на распределение полученных значений.


```python
sns.set_style('dark')
user_activity['session_duration'] = (user_activity['end'] - user_activity['start']).dt.seconds

user_activity['session_duration'].hist(bins=70, range=(0,4000))
```

Посмотрим на статистику

```python
user_activity['session_duration'].describe()
```

Стандартное отклонение очень большое – 1000 секунд (или около 17 минут), что означает, что продолжительность сессий распределена неравномерно. Также, если верить данным, кто-то провёл на Афише без малого сутки – 84480 секунд.

```python
print('Средняя продолжительность сессии – {} секунд'.format(int(user_activity['session_duration'].mode())))
```

### 1.4  Retention Rate / Как часто люди возвращаются?

Даты первых сеансов уже определены; можно присваивать сроки жизни когорт. За единицу измерения возьмём недели (для начала).

```python
user_activity['cohort_lifetime_weeks'] = user_activity['session_week'] - user_activity['first_session_week']
user_activity['cohort_lifetime_weeks'] = user_activity['cohort_lifetime_weeks'] / np.timedelta64(1,'W')
user_activity['cohort_lifetime_weeks'] = user_activity['cohort_lifetime_weeks'].astype(int)
```

Сгрупируем данные по неделям первых сеансов и сохраним их в отдельную переменную:

```python
cohorts_weeks = user_activity.groupby(['first_session_week','cohort_lifetime_weeks']).agg(
    {'id':'nunique'}).reset_index()
```

Добавим количество пользователей в когортах и сосчитаем retention rate:

```python
initial_users_count = cohorts_weeks[cohorts_weeks['cohort_lifetime_weeks'] == 0][['first_session_week','id']]
initial_users_count = initial_users_count.rename(columns={'id':'cohort_users'})
cohorts_weeks = cohorts_weeks.merge(initial_users_count,on='first_session_week')
cohorts_weeks['retention'] = cohorts_weeks['id']/cohorts_weeks['cohort_users']
```

Сделаем сводную таблицу:


```python
retention_pivot_weeks = cohorts_weeks.pivot_table(
    index='first_session_week',
    columns='cohort_lifetime_weeks',
    values='retention',
    aggfunc='sum')
```

```python
retention_pivot_weeks.head()
```

Сводная таблица огромная и непонятная. Сделаем тепловую карту:

```python
sns.set_style("white")
plt.figure(figsize=(20, 9), dpi=120)
plt.title('User Retention', y=1.02, fontsize=14)
sns.heatmap(retention_pivot_weeks.loc[:,1:], 
            cmap="Blues")
```

Тепловая карта лучше, но ненамного. Возможно, стоит выбрать месяцы как единицу измерения для времени жизни когорты. Попробуем:

```python
user_activity['cohort_lifetime_months'] = user_activity['session_month'] - user_activity['first_session_month']
user_activity['cohort_lifetime_months'] = user_activity['cohort_lifetime_months'] / np.timedelta64(1,'M')
user_activity['cohort_lifetime_months'] = user_activity['cohort_lifetime_months'].astype(int)
```

```python
cohorts_months = user_activity.groupby(
    ['first_session_month','session_month','cohort_lifetime_months']).agg({'id':'nunique'}).reset_index()
```

```python
cohorts_months
```

При таком преобразовании некоторым месяцам присываиваются неверные сроки жизни. Нужно исправить:

```python
for i in cohorts_months.index[0:len(cohorts_months)-1]:
    if cohorts_months['cohort_lifetime_months'][i] == cohorts_months['cohort_lifetime_months'][i+1]:
        cohorts_months['cohort_lifetime_months'][i+1]+=1
```

Также как и с неделями, находим количество пользователей в месячных когортах и считаем retention rate:


```python
initial_users_count = cohorts_months[cohorts_months['cohort_lifetime_months'] == 0][['first_session_month','id']]
initial_users_count = initial_users_count.rename(columns={'id':'cohort_users'})
cohorts_months = cohorts_months.merge(initial_users_count,on='first_session_month')
cohorts_months['retention'] = cohorts_months['id']/cohorts_months['cohort_users']
```

```python
retention_pivot_months = cohorts_months.pivot_table(
    index='first_session_month',
    columns='cohort_lifetime_months',
    values='retention',
    aggfunc='sum')

retention_pivot_months.loc['sum'] = retention_pivot_months.mean(axis=0)
```

Создадим переменную, в которой будут храниться названия месяцев – это для оси y в тепловых картах.

```python
labels = [calendar.month_abbr[i] + ' 2017' for i in range(6,13)] + [calendar.month_abbr[i] + ' 2018' for i in range(1,6)]
```

```python
sns.set_style("white")

plt.figure(figsize=(17,8),dpi=200)
plt.title('User Retention', y=1.02, fontsize=14)
sns.heatmap(retention_pivot_months.loc[:,1:], 
            annot=True, 
            fmt='.1%', 
            cmap="Blues",
            yticklabels=labels[:-1]+['Mean'])
```

Теперь тепловая карта выглядит информативнее. До второго месяца в среднем доходит только 6.5% пользователей, попробовавших сервис. Выделяется когорта сентября 2017 – там retention rate в первом месяце равен 8.5%. Меньше всего пользователей вернулось в когорте марта 2018 – 4.2%, а самыми верными оказались пользователи, пришедшие в июне 2017 – больше 4% пользователей этой когорты использовали Афишу на 9, 10, 11 и 12 месяце после первого перехода.

## 5 Средняя выручка с пользователя

Разделим суммарную выручку когорту на количество пользователей в ней и получим среднюю выручку с пользователя


```python
orders_grouped_by_cohorts['revenue_per_user'] = orders_grouped_by_cohorts['revenue']/orders_grouped_by_cohorts['uid']

mean_revenue_pivot  = orders_grouped_by_cohorts.pivot_table(
    index='first_order_month',
    columns='order_month',
    values='revenue_per_user', 
    aggfunc='mean')

mean_revenue_pivot['mean'] = mean_revenue_pivot.mean(axis=1)
```

```python
sns.set_style("white")
plt.figure(figsize=(20, 9), dpi=120)
plt.title('Mean Revenue', y=1.02, fontsize=14)
sns.heatmap(mean_revenue_pivot, 
            cmap="Blues", 
            annot=True,
            fmt='.0f',
            xticklabels=labels+['Сohort mean'], 
            yticklabels=labels)
```

Опять пользователи сентябрьской когорты выделяются на общем фоне.

Самая большая средняя выручка приходит с пользователей когорты декабря 2017, сентября 2017 и июня 2017. При этом, периоды активности различаются от когорты к когорте. Так, например, пользователи когорты декабря были наиболее расточительны в период с февраля по май, а пользователи июньской когорты больше тратились в сентябре.

Далее посчитаем сроки жизни когорт и добавим их в таблицу по заказам


```python
orders['cohort_lifetime_months'] = orders['order_month'] - orders['first_order_month']
orders['cohort_lifetime_months'] = orders['cohort_lifetime_months'] / np.timedelta64(1,'M')
orders['cohort_lifetime_months'] = orders['cohort_lifetime_months'].astype(int)
```

Объединим таблицы по заказам и расходам в одну по общей колонке с месяцами и сохраним результат в новую переменную

```python
revenue_and_costs = orders.merge(
    costs.groupby('month')['costs'].sum().reset_index(), 
    left_on='order_month', 
    right_on='month').groupby(
    ['first_order_month', 'order_month','cohort_lifetime_months','costs']).agg({'uid':'nunique','revenue':'sum'}).reset_index()

revenue_and_costs = revenue_and_costs.merge(orders.groupby(['first_order_month']).agg({'uid':'nunique'}), 
                        on='first_order_month').rename(columns={'uid_x':'monthly_cohort_users','uid_y':'total_cohort_users'})

revenue_and_costs = revenue_and_costs.merge(
    revenue_and_costs.groupby('first_order_month')['revenue'].sum(), 
    on='first_order_month').rename(columns={'revenue_x':'monthly_cohort_revenue','revenue_y':'total_cohort_revenue'})

revenue_and_costs.drop(['costs'], axis=1).merge(
    revenue_and_costs[['order_month','costs']], 
    left_on='first_order_month', 
    right_on='order_month').drop(['order_month_y'], axis=1).rename(columns={'order_month_x':'order_month'})

for i in revenue_and_costs.index[0:len(cohorts_months)-1]:
    if revenue_and_costs['cohort_lifetime_months'][i] == revenue_and_costs['cohort_lifetime_months'][i+1]:
        revenue_and_costs['cohort_lifetime_months'][i+1]+=1
```

Посчитаем стоимость привлечения клиента (CAC), объём выручки с пользователей (LTV) за шесть месяцев (уже семь когорт достигли этого возраста) и возврат маркетинговых инвестиций (ROMI)

Ограничим выборку только когортами старше января 2018

```python
revenue_and_costs = revenue_and_costs[revenue_and_costs['first_order_month']<'2018-01-01'] 

revenue_and_costs['cac'] = revenue_and_costs['costs'] / revenue_and_costs['total_cohort_users']
revenue_and_costs['ltv'] = revenue_and_costs['monthly_cohort_revenue'] / revenue_and_costs['total_cohort_users']
```

```python
output = revenue_and_costs[revenue_and_costs['cohort_lifetime_months'] < 6].pivot_table(
        index='first_order_month', 
        columns='cohort_lifetime_months', 
        values='ltv', 
        aggfunc='mean').round(2)

output['total'] = output.sum(axis=1)

sns.set_style("white")
plt.figure(figsize=(20, 9), dpi=120)
plt.title('LTV', y=1.02, fontsize=14)
sns.heatmap(output.round(2), 
            annot=True, 
            cmap="Blues",
            yticklabels=[calendar.month_abbr[i] + ' 2017' for i in range(6,13)])
```

```python
mean_ltv = output['total'].mean()
print('Средний LTV равен', round(mean_ltv,2))
```

```python
revenue_and_costs['romi'] = revenue_and_costs['ltv'] / revenue_and_costs['cac']
```

Создадим сводную таблицу по ROMI и месяцам жизни месячных когорт

```python
output = revenue_and_costs[revenue_and_costs['cohort_lifetime_months'] < 6].pivot_table(
        index='first_order_month',
        columns='cohort_lifetime_months',
        values='romi',
        aggfunc='mean')
```

```python
sns.set_style("white")
plt.figure(figsize=(20, 9), dpi=120)
plt.title('ROMI', y=1.02, fontsize=14)
sns.heatmap(output.cumsum(axis=1).round(2), 
            annot=True,
            fmt='.0%',
            cmap="Blues",
            yticklabels=[calendar.month_abbr[i] + ' 2017' for i in range(6,13)])
```

Лучше всего стартовала когорта июля 2017 – 63% маркетинговых инвестиций вернулись уже в первый месяц, а самый высокий финальный показатель окупаемости показала когорта сентября 2017 – 106% к седьмому месяцу. Также практически окупилась когорта декабря 2017 (98% к седьмому месяцу). Самой неудачной оказалась когорта августа 2017, стартовавшая лишь с 49% ROMI в первом месяце и закончившая c 58% к концу выбранного периода.


```python
output.cumsum(axis=1).mean(axis=0)
```

В среднем когорты не успевают окупаться к седьмому месяцу – их показатель ROMI достигает только 80%

## LTV рекламных источников

Подготовим датафрейм с данными по выручке и затратам по источникам

```python
sales_source_total['date'] = pd.to_datetime(sales_source_total['date'])

sources_revenue_and_costs = sales_source_total.merge(
    costs[['source_id','costs','dt','month']], 
    left_on=['date','source id'], 
    right_on=['dt','source_id']).drop(
    ['dt','source_id','device'],axis=1).dropna()

sources_list = [1,2,3,4,5,9,10]
```

Метод для составления датафреймов по отдельным источникам


```python
def create_source_dataframe(source_id):
    first_order_dates = sources_revenue_and_costs[sources_revenue_and_costs['source id'] == source_id].groupby('uid')['month'].min()
    first_order_dates.name = 'first_month'
    source = sources_revenue_and_costs[sources_revenue_and_costs['source id'] == source_id].merge(first_order_dates, on='uid')
    
    source['cohort_age'] = source['month'] - source['first_month']
    source['cohort_age'] = source['cohort_age'] / np.timedelta64(1,'M')
    source['cohort_age'] = source['cohort_age'].astype(int)

    source = source.groupby(['first_month','cohort_age']).agg(
        {'uid':'nunique',
         'revenue':'sum',
         'costs':'sum'}).reset_index()

    users_total = sales_source_total[sales_source_total['source id'] == source_id]['uid'].nunique()

    source['cac'] = source['costs'] / users_total
    source['ltv'] = source['revenue'] / users_total
    source['romi'] = source['ltv'] / source['cac']
    
    return source
```

Метод для вычисления ROMI средней когорты источника


```python
def mean_cohort_romi(source_id):
    output = create_source_dataframe(source_id).query('first_month < "2018-01-01" and cohort_age < 6').pivot_table(
            index='first_month', 
            columns='cohort_age', 
            values='romi', 
            aggfunc='mean').round(2)
    
    mean_romi = output.cumsum(axis=1).mean(axis=0)
    mean_romi.name = 'source_{}'.format(source_id)
    
    return mean_romi.reset_index().set_index('cohort_age').T
```

Например

```python
mean_cohort_romi(source_id=10)
```

Составим таблицу ROMI средних когорт по источникам


```python
romi_pivot = pd.concat([mean_cohort_romi(i) for i in sources_list])
```

Нарисуем тепловую карту

```python
plt.figure(figsize=(20, 9), dpi=120)
plt.title('ROMI', y=1.02, fontsize=14)
sns.heatmap(romi_pivot, cmap="Blues", annot=True, fmt='.0%')
```

К концу шестого месяца окупились только источники 1 и 10, а источник 9 окупился на 94%. Источники 4,5, и особенно 3, оказались убыточными.
